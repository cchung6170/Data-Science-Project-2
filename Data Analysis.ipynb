```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Merged Video Game Sales and Reviews.csv')
df["user score"] = pd.to_numeric(df["user score"], errors="coerce")
df
```

```python
# Setting up a list of the sales columns to loop create scatterplots
sales_cols = ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]
```

```python
# Scatter plots based off of critics reviews
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, sales_col in enumerate(sales_cols):
    plot_df = df[["score", sales_col]].dropna()
    plot_df = plot_df[plot_df[sales_col] > 0]

    axes[i].scatter(plot_df["score"], plot_df[sales_col], alpha=0.5)
    axes[i].set_xlabel("Critic Review Score")
    axes[i].set_ylabel(sales_col)
    axes[i].set_title(f"Critic Score vs. {sales_col}")

axes[-1].set_visible(False)

fig.suptitle("Critic Review Score vs. Video Game Sales (millions)", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, sales_col in enumerate(sales_cols):
    plot_df = df[["user score", sales_col]].dropna()
    plot_df = plot_df[plot_df[sales_col] > 0]

    axes[i].scatter(plot_df["user score"], plot_df[sales_col], alpha=0.5)
    axes[i].set_xlabel("User Review Score")
    axes[i].set_ylabel(sales_col)
    axes[i].set_title(f"User Score vs. {sales_col}")

    axes[i].set_xticks(np.arange(0, 11, 1))

axes[-1].set_visible(False)

fig.suptitle("User Review Score vs. Video Game Sales (millions)", fontsize=16)
plt.tight_layout()
plt.show()
```
